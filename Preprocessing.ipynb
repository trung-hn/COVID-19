```python
import pandas as pd
from Bio import SeqIO
from Bio.SeqRecord import SeqRecord
import collections
```

```python
pd.set_option('display.max_rows', 20)
```

```python
df = pd.read_csv("./raw_data/metadata.tsv", sep="\t")
# Filter by conditions
df = df[(df["host"] == 'Human') & (df["age"] != "?") & ((df["sex"] == "Male") | (df["sex"] == "Female"))]
# Only want a few columns
df = df[['gisaid_epi_isl', "country", "age", "sex", "date_submitted"]]
# Preview
df
```

```python
data_set = collections.defaultdict(list)
for i, (gisaid, *rest) in df.iterrows():
    # rest = (country, age, sex, date), in that order
    data_set[gisaid].extend(rest) 
# Example of dataset:
for i, val in enumerate(data_set.items()):
    print(val)
    if i == 10: break
```

### Filter faster file based those available in data_set

```python
records = []
for i, record in enumerate(SeqIO.parse("D:\gisaid_hcov-19_2020_05_14_18.fasta", "fasta")):
    header = record.id.split("|")
    if len(header) > 1 and header[1] in data_set:
        records.append(record)
```

### Save new fasta file

```python
SeqIO.write(records, "./data/gisaid_cov2020_sequences_filtered.fasta", "fasta")
```
